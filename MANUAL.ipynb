```yaml

    metadata_figure_settings:
      style: 
        placement: caption
      license:
        summaries: true
        individual: true
```

### Example 11b: Automatic BibTeX Generation

When `generate_bib: true` is set in the configuration and you specify a `:bib:` key that doesn't exist in any `.bib` file, the extension will automatically generate a BibTeX entry from the figure's metadata options.

**Configuration required:**

```yaml
sphinx:
  config:
    metadata_figure_settings:
      bib:
        extract_metadata: true
        generate_bib: true
        output_file: "_generated_figures.bib"
```

**Usage:**

````md
```{figure} /images/TeachBooks_logo.svg
:name: tb_logo_metadata11b
:width: 50%
:bib: AutoGeneratedKey
:author: TeachBooks Team
:date: 2024-06-15
:license: CC BY
:source: https://github.com/TeachBooks/manual

This figure will auto-generate a BibTeX entry.
```

Citation: {cite:p}`AutoGeneratedKey`
````

This will automatically create an entry in `_generated_figures.bib`:

````bibtex
@misc{AutoGeneratedKey,
  author = {TeachBooks Team},
  title = {This figure will auto-generate a BibTeX entry.},
  year = {2024},
  date = {2024-06-15},
  url = {https://github.com/TeachBooks/manual},
  howpublished = {\url{https://github.com/TeachBooks/manual}},
  note = {License: CC BY},
}
````

**Key features:**
- The entry is generated during the `config-inited` phase, **before** sphinxcontrib-bibtex loads the `.bib` files
- This means the citation works in a **single build** - no need to build twice
- If a key already exists in any `.bib` file, it won't be overwritten
- The figure caption is used as the BibTeX `title` field
- Supports multiple directive formats: backticks (` ``` `), colons (`:::`), and RST (`.. figure::`)
